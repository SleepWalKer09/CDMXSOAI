# Instalacion de Tensorflow

## Desde las fuentes

Para Deep Learning (por cuestiones de velocidad y tiempo) es necesario hacer uso de la tarjeta de video y de la libreria CUDA de NVIDIA
***APPLE no tiene soporte CUDA desde tensorflow 1.2*** 
***Solo los productos de NVIDIA son compatibles con Tensorflow***

## Instalacion de CUDA

Para instalar CUDA en Linux tenemos que seguir los siguientes pasos:

    sudo apt-get update 
    sudo apt-get upgrade

El siguiente comando nos debe arrojar la tarjeta de video que tenemos en nuestra maquina y tenemos que asegurarnos de que este listada como
compatible con CUDA, si no esta explicitamente listada no es posible utilizar CUDA
    
    lspci | grep -i nvidia

Podemos checar la lista de tarjetas de video compatibles con CUDA aqui:

https://developer.nvidia.com/cuda-gpus

Tenemos que verificar que la version de Linux que tenemos es compatible con CUDA con este comando:

    uname -m && cat /etc/*release
    
Despues instalamos las dependencias necesarias con estos comandos:

    sudo apt-get install build-essential 
    sudo apt-get install cmake git unzip zip
    sudo add-apt-repository ppa:deadsnakes/ppa
    sudo apt-get update
    sudo apt-get install python2.7-dev python3.5-dev python3.6-dev pylint
    
Tenemos que ver cual que Kernel tenemos con el siguiente comando:

    uname -r
    
*Si tenemos un kernel viejo(poco probable si actualizamos Ubuntu continuamente) lo actualizamos con el siguiente comando:

    sudo apt-get install linux-headers-$(uname -r)
    

Para usar Tensorflow 1.8 tenemos que usar versiones actuales de CUDA(9.2 en nuestro caso) y por lo tanto es necesario quitar las versiones 
anteriores que pudieran estar instaladas con los siguientes comandos:

    sudo apt-get purge nvidia*
    sudo apt-get autoremove
    sudo apt-get autoclean
    sudo rm -rf /usr/local/cuda*
    
Al dia de hoy(23 de Agosto 2018) instalamos CUDA con los siguientes comandos:(En fechas posteriores esto puede cambiar):


    sudo apt-key adv --fetch-keys http://developer.download.nvidia.com/compute/cuda/repos/ubuntu1710/x86_64/7fa2af80.pub
    
    echo "deb https://developer.download.nvidia.com/compute/cuda/repos/ubuntu1710/x86_64 /" | sudo tee /etc/apt/sources.list.d/cuda.list
    
    sudo apt-get update 
    sudo apt-get -o Dpkg::Options::="--force-overwrite" install cuda-9-2 cuda-drivers
   
   
Ahora tenemos que agregar CUDA a nuestro PATH con el siguiente comando:

    echo 'export PATH=/usr/local/cuda-9.2/bin${PATH:+:${PATH}}' >> ~/.bashrc
    
    echo 'export LD_LIBRARY_PATH=/usr/local/cuda-9.2/lib64${LD_LIBRARY_PATH:+:${LD_LIBRARY_PATH}}' >> ~/.bashrc

Despues tenemos que actualizar nuestro bash con los siguientes comandos

    source ~/.bashrc
    sudo ldconfig
    

Con el siguiente comando verificamos que se han instalado correctamente los controladores de nuestra tarjeta NVIDIA

    nvidia-smi

Con el siguiente comando verificamos que se han instalado correctamente los controladores de CUDA

    nvcc --version

***Si los comandos no funcionan o arrojan error reinicia la maquina***

Para hacer uso de CUDA y Tensorflow es necesario instalar cuDNN para CUDA 9.2, los podemos bajar aqui 
(Es necesario tener una cuenta de NVIDIA Developer):

https://developer.nvidia.com/cudnn 

Una vez que bajamos el archivo adecuado vamos a nuestra carpeta de Downloads(cd Downloads/) y ejecutamos los siguientes comandos:

    tar -xf cudnn-9.2-versiondescargada.tgz
    sudo cp -R cuda/include/* /usr/local/cuda-9.2/include
    sudo cp -R cuda/lib64/* /usr/local/cuda-9.2/lib64
    
De igual manera vamos a necesitar la libreria de NCCL para utlizar versiones de CUDA arriba de 9.0 
(Es necesario tener una cuenta de NVIDIA Developer):

***Al dia de hoy (15 de Agosto de 2018) es necesario bajar la version O/S agnostic para que funcione bien.

https://developer.nvidia.com/nccl

tar -xf versionnccl_x86_64.txz
cd carpetanccl_x86_64
sudo cp -R * /usr/local/cuda-9.2/targets/x86_64-linux/
sudo ldconfig

Despues tenemos que instalar las dependencias necesarias con el siguiente comando:

    sudo apt-get install libcupti-dev
    echo 'export LD_LIBRARY_PATH=/usr/local/cuda/extras/CUPTI/lib64:$LD_LIBRARY_PATH' >> ~/.bashrc

Ahora instalamos las siguientes librerias:

    sudo apt-get install python3-numpy python3-dev python3-pip python3-wheel

Para construir Tensorflow desde las fuentes es necesario instalar Bazel

    cd Downloads/
    wget https://github.com/bazelbuild/bazel/releases/download/0.14.0/bazel-0.14.0-installer-linux-x86_64.sh
    chmod +x bazel-0.14.0-installer-linux-x86_64.sh
    ./bazel-0.14.0-installer-linux-x86_64.sh --user
    echo 'export PATH="$PATH:$HOME/bin"' >> ~/.bashrc

Tenemos que hacer refresh de nuestro archivo bash con el siguiente comando:

    source ~/.bashrc
    sudo ldconfig

Ahora tenemos que hacer el pull del repo de Tensorflow con los siguientes comandos:

    cd Downloads
    git clone https://github.com/tensorflow/tensorflow.git
    cd tensorflow
    git pull
    git checkout r1.8

Con el siguiente comando vamos a customizar Tensorflow para nuestra tajeta Nvidia

Una vez que terminanos las configuracion de Tensorflow tenemos que compilarlo(TARDA MAS DE 2 horas) con el siguiente comando:

    bazel build --config=opt --local_resources 2048,.5,1.0 --config=mkl //tensorflow/tools/pip_package:build_pip_package

Una vez que este proceso termina tenemos que construir el paquete whl para poder instalarlo con el siguiente comando:

    bazel-bin/tensorflow/tools/pip_package/build_pip_package tensorflow_pkg

Una vez que tenemos nuestro archivo whl lo podemos instalar en un ambiente virtual 
***Ojo se tiene que instalar en un ambiente virtual forzosamente para no alterar el OS***:

    sudo apt-get install virtualenv
    virtualenv nombredelambiente -p /usr/bin/python3
    source nombredelambiente/bin/activate
    pip3 install tensorflow*.whl

Vamos a comprobar nuestra instalacion de Tensorflow con los siguientes comandos:

    python
    import tensorflow as tf
    hello = tf.constant('Hello, TensorFlow!')
    sess = tf.Session()
    print(sess.run(hello))

Nos debe arrojar el mensaje: Hello, Tensorflow!


## Desde PIP

El paquete de pip de Tensorflow para GPU esta compilado para funcionar con CUDA 9.0, por lo que es necesario seguir todos los pasos anteriores para la instalacion de CUDA pero seleccionando la opcion 9.0 en lugar de la 9.2

Una vez instalada la version 9.0 de CUDA simplemente vamos a nuestro ambiente virtual e introducimos:

    pip3 install tensorflow-gpu==1.8.0
    
Una vez terminanda la instalacion comprobamos la instalacion con los siguientes comandos:

    python
    import tensorflow as tf
    hello = tf.constant('Hello, TensorFlow!')
    sess = tf.Session()
    print(sess.run(hello))


Nos debe arrojar el mensaje: Hello, Tensorflow!
